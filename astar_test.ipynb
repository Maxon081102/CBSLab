```python
from astar import astar, SearchTreePQS, Node
from Constraints import Constraints
from CBS_Agent import BaseNode, Agent
from Solutions import Solution, Solutions
from Solutions import make_path

from draw import draw, draw_dynamic
from map import Map
```

```python
def manhattan(x1, y1, x2, y2):
    return abs(x1 - x2) + abs(y1 - y2)


my_map = """
. # . # . # . #
. . . # . . . .
. # . # . . # .
. # . . . # . .
. . # # . . # .
# . . . . . . .
. . . . # # # .
# # # . . . . .
"""
w, h = 8, 8

agent = Agent(0, (0, 0), (7, 7))
constraints = Constraints()
constraints.add_constraint(agent, 20, BaseNode(1, 6))
constraints.add_constraint(agent, 35, BaseNode(0, 2))
constraints.add_constraint(agent, 50, BaseNode(7, 6))



test_map = Map()
test_map.read_from_string(my_map, w, h) 

found, end_node, steps = astar(test_map, agent, constraints, heuristic_func=manhattan, search_tree=SearchTreePQS)
sol = Solutions()
sol.add_solution(found, end_node, steps)
# path, _ = make_path(end_node)
# draw(test_map, Node(0, 0), Node(7, 7), path)
draw_dynamic(test_map, sol)
# print(len(path))
```
